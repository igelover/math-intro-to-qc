# Pauli Operators

_prepared by Israel Gelover_

### <a name="definition_2_28">Definition 2.28</a>

We define _Pauli Operators_ as

\begin{equation*}
    \begin{split}
        \sigma_0:\mathbb{B} \to \mathbb{B} \enspace &, \enspace \sigma_0\ket{0} = \ket{0}, \sigma_0\ket{1} = \ket{1} \\
        \sigma_x:\mathbb{B} \to \mathbb{B} \enspace &, \enspace \sigma_x\ket{0} = \ket{1}, \sigma_x\ket{1} = \ket{0} \\
        \sigma_y:\mathbb{B} \to \mathbb{B} \enspace &, \enspace \sigma_y\ket{0} = i\ket{1}, \sigma_x\ket{1} = -i\ket{0} \\
        \sigma_z:\mathbb{B} \to \mathbb{B} \enspace &, \enspace \sigma_z\ket{0} = \ket{0}, \sigma_x\ket{1} = -\ket{1}
    \end{split}
\end{equation*}

where $\mathbb{B}$ is the <a href="./BlochSphere.ipynb#definition_2_7">Bloch Sphere</a>.  
Note that these operators are unitary and self-adjoint.

Given this definition of how they act on the elements of the computational base, we can see them as external products, considering that in the braket, the bra represents the input and the ket is the result of applying the operator, in this way

\begin{equation*}
    \begin{split}
        \sigma_0 &= \ketbra{0}{0} + \ketbra{1}{1} \\
        \sigma_x &= \ketbra{1}{0} + \ketbra{0}{1} \\
        \sigma_y &= i\ketbra{1}{0} - i\ketbra{0}{1} \\
        \sigma_z &= \ketbra{0}{0} - \ketbra{1}{1}
    \end{split}
\end{equation*}

From this representation of exterior products, we can see them as matrices (in the computational base) using the method described <a href="./SpectralTheory.ipynb#examples">here</a>

\begin{equation*}
    \sigma_0 = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix}, \enspace \sigma_x = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix}, \enspace \sigma_y = \begin{pmatrix} 0 & -i \\ i & 0 \end{pmatrix}, \enspace \sigma_z = \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix}
\end{equation*}

### <a name="proposition_2_29">Proposition 2.29</a>

Let $\ket{\psi} \in \mathbb{B}$, we can express it in its general form as

\begin{equation*}
    \ket{\psi} = \cos \left(\frac{\theta}{2}\right)\ket{0} + e^{i\phi}\sin \left(\frac{\theta}{2}\right)\ket{1}
\end{equation*}

Then

1. $\expected{\sigma_x}_\psi = \sin \theta \cos \phi = x$
2. $\expected{\sigma_y}_\psi = \sin \theta \sin \phi = y$
3. $\expected{\sigma_z}_\psi = \cos \theta = z$

**Proof:**

1. Let's see first that

\begin{equation*}
    \sigma_x\ket{\psi} = e^{i\phi}\sin\left(\frac{\theta}{2}\right)\ket{0} + \cos\left(\frac{\theta}{2}\right)\ket{1}
\end{equation*}
$\implies$
\begin{equation*}
    \begin{split}
        \bra{\psi}\sigma_x\ket{\psi} &= \left(\cos\left(\frac{\theta}{2}\right)\bra{0} + e^{-i\phi}\sin\left(\frac{\theta}{2}\right)\bra{1}\right) \left(e^{i\phi}\sin\left(\frac{\theta}{2}\right)\ket{0} + \cos\left(\frac{\theta}{2}\right)\ket{1}\right) \\
        &= e^{i\phi}\sin\left(\frac{\theta}{2}\right)\cos\left(\frac{\theta}{2}\right)\braket{0}{0} + \cos^2\left(\frac{\theta}{2}\right)\braket{0}{1} \\
        &+ \sin^2\left(\frac{\theta}{2}\right)\braket{1}{0} + e^{-i\phi}\sin\left(\frac{\theta}{2}\right)\cos\left(\frac{\theta}{2}\right)\braket{1}{1} \\
        &= e^{i\phi}\frac{1}{2}\sin(\theta) + e^{-i\phi}\frac{1}{2}\sin(\theta)
    \end{split}
\end{equation*}

Recall that $\sin(2x) = 2\sin(x)\cos(x)$, substituting $x=\frac{\theta}{2}$ we get the last equality. On the other hand we also know that $e^{i\phi} = \cos(\phi) + i\sin(\phi)$, in this way we obtain

\begin{equation*}
    \begin{split}
        \bra{\psi}\sigma_x\ket{\psi} &= \frac{1}{2}\sin(\theta)(e^{i\phi} + e^{-i\phi}) \\
        &= \frac{1}{2}\sin(\theta)(\cos(\phi) + i\sin(\phi) + \cos(\phi) - i\sin(\phi)) \\
        &= \frac{1}{2}\sin(\theta)2\cos(\phi) = \sin(\theta)\cos(\phi)
    \end{split}
\end{equation*}
$\therefore \expected{\sigma_x}_\psi = \sin(\theta)\cos(\phi)$

2. Let's see first that

\begin{equation*}
    \sigma_y\ket{\psi} = -ie^{i\phi}\sin\left(\frac{\theta}{2}\right)\ket{0} + i\cos\left(\frac{\theta}{2}\right)\ket{1}
\end{equation*}
$\implies$
\begin{equation*}
    \begin{split}
        \bra{\psi}\sigma_y\ket{\psi} &= \left(\cos\left(\frac{\theta}{2}\right)\bra{0} + e^{-i\phi}\sin\left(\frac{\theta}{2}\right)\bra{1}\right) \left(-ie^{i\phi}\sin\left(\frac{\theta}{2}\right)\ket{0} + i\cos\left(\frac{\theta}{2}\right)\ket{1}\right) \\
        &= -ie^{i\phi}\sin\left(\frac{\theta}{2}\right)\cos\left(\frac{\theta}{2}\right)\braket{0}{0} + ie^{-i\phi}\sin\left(\frac{\theta}{2}\right)\cos\left(\frac{\theta}{2}\right)\braket{1}{1} \\
        &= \frac{1}{2}\sin(\theta)i\left(-e^{i\phi} + ie^{-i\phi}\right) \\
        &= \frac{1}{2}\sin(\theta)i(-\cos(\phi) - i\sin(\phi) + \cos(\phi) - i\sin(\phi)) \\ 
        &= \frac{1}{2}\sin(\theta)(i)(-i)(2\sin(\phi)) = \sin(\theta)\sin(\phi)
    \end{split}
\end{equation*}

Since $\braket{0}{1} = 0$, $\braket{1}{0} = 0$ y $e^{i\phi} = \cos(\phi) + i\sin(\phi)$  
$\therefore \expected{\sigma_y}_\psi = \sin(\theta)\sin(\phi)$

3. Let's see first that

\begin{equation*}
    \sigma_z\ket{\psi} = \cos\left(\frac{\theta}{2}\right)\ket{0} - e^{i\phi}\sin\left(\frac{\theta}{2}\right)\ket{1}
\end{equation*}
$\implies$
\begin{equation*}
    \begin{split}
        \bra{\psi}\sigma_z\ket{\psi} &= \left(\cos\left(\frac{\theta}{2}\right)\bra{0} + e^{-i\phi}\sin\left(\frac{\theta}{2}\right)\bra{1}\right) \left(\cos\left(\frac{\theta}{2}\right)\ket{0} - e^{i\phi}\sin\left(\frac{\theta}{2}\right)\ket{1}\right) \\
        &= \cos^2\left(\frac{\theta}{2}\right)\braket{0}{0} - \sin^2\left(\frac{\theta}{2}\right)\braket{1}{1} \\
        &= \cos(\theta)
    \end{split}
\end{equation*}

Since $\braket{0}{1} = 0$, $\braket{1}{0} = 0$ y $\cos(2x) = \cos^2(x) - \sin^2(x)$, substituting $x = \frac{\theta}{2}$ we get the last equality.  
$\therefore \expected{\sigma_z}_\psi = \cos(\theta)$
