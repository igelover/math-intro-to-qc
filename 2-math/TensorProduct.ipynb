### <a name="definition_2_9">Definition 2.9</a> Tensor Product or Kronecker Product

Let $V$ be a space of dimension $n$ and $W$ a space of dimension $m$, and $\set{\ket{v_j}}_{j=1}^{n}$, $\set{\ket{w_k}}_{k=1}^{m}$ bases of $V$ and $W$ respectively. That is,

\begin{equation*}
    \begin{split}
        \ket{v} \in V &\implies \ket{v} = a_1\ket{v_1} + a_2\ket{v_2} + \cdots + a_n\ket{v_n} \enspace \text{ i.e. } \ket{v} = \begin{pmatrix} a_1 \\ a_2 \\ \vdots \\ a_n \end{pmatrix} \\
        \ket{w} \in W &\implies \ket{w} = b_1\ket{w_1} + b_2\ket{w_2} + \cdots + b_m\ket{w_m} \enspace \text{ i.e. } \ket{w} = \begin{pmatrix} b_1 \\ b_2 \\ \vdots \\ b_m \end{pmatrix}
    \end{split}
\end{equation*}

We define the tensor product of $\ket{v}$ and $\ket{w}$ as

\begin{equation*}
    \ket{v} \otimes \ket{w} = \begin{pmatrix} a_1 \\ a_2 \\ \vdots \\ a_n \end{pmatrix} \otimes \begin{pmatrix} b_1 \\ b_2 \\ \vdots \\ b_m \end{pmatrix} = \begin{pmatrix} a_1 \begin{pmatrix} b_1 \\ b_2 \\ \vdots \\ b_m \end{pmatrix} \\ \vdots \\ a_n \begin{pmatrix} b_1 \\ b_2 \\ \vdots \\ b_m \end{pmatrix} \end{pmatrix} = \begin{pmatrix} a_1 b_1 \\ a_1 b_2 \\ \vdots \\ a_1 b_m \\ \vdots \\ a_n b_1 \\ a_n b_2 \\ \vdots \\ a_n b_m \end{pmatrix} = \ket{v,w} = \ket{vw}
\end{equation*}

In a similar fashion, we define the tensor product of $V$ and $W$ as

\begin{equation*}
    V \otimes W = \angleset{\set{\ket{v_j} \otimes \ket{w_k} \mid k = 1,2,\cdots,m \enspace j = 1,2,\cdots,n}}
\end{equation*}

Where $V \otimes W$ is the space generated by the tensor products of the elements of the bases of $V$ and $W$ respectively. The dimension of this new space is $n \times m$.

### <a name="remark_2_10">Remark 2.10</a>

1. Let $V$ and $W$ be spaces with inner product $\cdot_V$ and $\cdot_W$ respectively, then

\begin{equation*}
    (\ket{v_1} \otimes \ket{w_1}) \cdot_{V \otimes W} (\ket{v_2} \otimes \ket{w_2}) = (\ket{v_1} \cdot_V \ket{v_2}) (\ket{w_1} \cdot_W \ket{w_2})
\end{equation*}

is an inner product in $V \otimes W$. That is, we can construct an inner product in $V \otimes W$ through the internal products defined in each space.

2. $\forall \ket{v_1}, \ket{v_2} \in V, \ket{w_1}, \ket{w_2} \in W, a \in \mathbb{C}$, we have that
    - $a(\ket{v_1} \otimes \ket{w_1}) = (a\ket{v_1}) \otimes \ket{w_1} = \ket{v_1} \otimes (a\ket{w_1})$
    - $(\ket{v_1} + \ket{v_2}) \otimes \ket{w_1} = (\ket{v_1} \otimes \ket{w_1}) + (\ket{v_2} \otimes \ket{w_1})$
    - $\ket{v_1} \otimes (\ket{w_1} + \ket{w_2}) = (\ket{v_1} \otimes \ket{w_1}) + (\ket{v_1} \otimes \ket{w_2})$

### <a name="definition_2_11">Definition 2.11</a>

Let $V_1, V_2, W_1, W_2$ be vector spaces.  
Given $\hat{A}:V_1 \to V_2$ y $\hat{B}:W_1 \to W_2$ linear operators, we define

\begin{equation*}
    \hat{A} \otimes \hat{B}:V_1 \otimes W_1 \to V_2 \otimes W_2
\end{equation*}

as the linear operator such that

\begin{equation*}
    \hat{A} \otimes \hat{B}(\ket{a} \otimes \ket{b}) = \left(\hat{A}\ket{a}\right) \otimes \left(\hat{B}\ket{b}\right)
\end{equation*}
