# Heisenberg Principle

_prepared by Israel Gelover_

What is known as the _Heisenberg Uncertainty Principle_ is a particular case of a more general theorem, in the sense that it applies to any pair of self-adjoint operators, while the Heisenberg Uncertainty Principle is limited to considering two observables that are position and momentum. Let's see what we need to understand these principles.

### <a name="definition_2_22">Definition 2.22</a> Expected Value

Let $A:\mathcal{H} \to \mathcal{H}$ be a linear operator, $\ket{\psi} \in \mathcal{H}$. We define the _Expected Value_ of $A$ with respect to $\ket{\psi}$ as:

\begin{equation*}
    \expected{A}_\psi = \bra{\psi} A \ket{\psi}
\end{equation*}

Please note that the expected value is a scalar, since from this definition we can see that we have a row vector multiplying a matrix and then multiplying a column vector, what we get as a result is a scalar.

The concept of expected value in Probability Theory that we know as the mean of a distribution, is related to this scalar that we are obtaining. The interpretation will be clearer when the postulates of quantum mechanics are listed, but for now we can say that if we see the ket $\ket{\psi}$ as a linear combination of the basis of a self-adjoint operator, we have a distribution of eigenvalues of the operator, and precisely what this scalar is giving us is the mean value of that distribution of eigenvalues.

### <a name="definition_2_23">Definition 2.23</a> Standard Deviation

Let $A:\mathcal{H} \to \mathcal{H}$ be a linear operator, $\ket{\psi} \in \mathcal{H}$. We define the Standard Deviation of $A$ with respect to $\ket{\psi}$ as:

\begin{equation*}
    \Delta_\psi A = \sqrt{\expected{A^2}_\psi - \expected{A}_\psi^2}
\end{equation*}

In the same way, this scalar is related to the concept of standard deviation that we know from probability theory and in fact, we define it in the same way as it is obtained in probability, that is, as a function of the expected value.

### <a name="definition_2_24">Definition 2.24</a> Commutator and Anticommutator

Let $A,B:\mathcal{H} \to \mathcal{H}$ be linear operators. We define

\begin{equation*}
    \begin{split}
        [A,B] &= AB - BA \enspace \longrightarrow \text{ Commutator} \\
        \left\{A,B\right\} &= AB + BA \enspace \longrightarrow \text{ Anticommutator}
    \end{split}
\end{equation*}

The commutator is a matrix, but we can think of it as a measure of how much operators $A$ and $B$ commute, in such a way that if we obtain matrix **0**, operators $A$ and $B$ commute.

### <a name="proposition_2_25">Proposition 2.25</a>

Let $A,B:\mathcal{H} \to \mathcal{H}$ be linear operators and $\ket{f} \in \mathcal{H}$ such that $\braket{f}{f} = 1$.  
If $H = A - \expected{A}_f I$ and $G = B - \expected{B}_f I$, then

1. $\expected{[H,G]}_f = \expected{[A,B]}_f$
2. $\expected{H^2}_f = (\Delta_f A)^2$
3. $\expected{G^2}_f = (\Delta_f B)^2$

**Proof:**

1. Let us note that

\begin{equation*}
    \begin{split}
        [H,G] &= HG-GH \\
        &= (A-\expected{A}_f I)(B-\expected{B}_f I) - (B-\expected{B}_f I)(A-\expected{A}_f I) \\
        &= AB - A\expected{B}_f I - \expected{A}_f IB + \expected{A}_f I\expected{B}_f I \\
        &- (BA - B\expected{A}_f I - \expected{B}_f IA + \expected{B}_f I\expected{A}_f I) \\
        &= AB-BA \\
        &= [A,B]
    \end{split}
\end{equation*}

Since $\expected{A}_f$ and $\expected{B}_f$ are scalars $\enspace \therefore \expected{[H,G]}_f = \expected{[A,B]}_f$

2. On the other hand we have that

\begin{equation*}
    \begin{split}
        \expected{H^2}_f &= \expected{(A-\expected{A}_f I)(A-\expected{A}_f I)}_f \\
        &= \expected{A^2 - A\expected{A}_f I - \expected{A}_f IA + (\expected{A}_f I)^2}_f \\
        &= \expected{A^2}_f - \expected{A}_f\expected{A}_f\expected{I}_f - \expected{A}_f\expected{I}_f\expected{A}_f + \expected{A}_f^2\expected{I^2}_f \\
        &= \expected{A^2}_f - 2\expected{A}_f^2 + \expected{A}_f^2 \\
        &= \expected{A^2}_f - \expected{A}_f^2
    \end{split}
\end{equation*}

Since, it is clear that $\expected{I}_f = 1$ and if $T$ is a constant operator, let's say $T = c$ then $\expected{T}_f = \bra{f}T\ket{f} = \bra{f}c\ket{f} = c$, because $\braket{f}{f} = 1$.

$\therefore \expected{H^2}_f = \expected{A^2}_f - \expected{A}_f^2 = (\Delta_f A)^2$

3. Doing a similar calculation, we can see that $\expected{G^2}_f = (\Delta_f B)^2$

### <a name="lemma_2_26">Lemma 2.26</a> Cauchy-Schwarz Inequality

$$
\abs{\braket{f}{g}}^2 \leq \braket{f}{f} \braket{g}{g} \enspace \forall \ket{f}, \ket{g} \in \mathcal{H}
$$

Note that this is the same inequality that we already know from Calculus, only written in Dirac notation.

### <a name="theorem_2_27">Theorem 2.27</a> Heisenberg Principle

Let $A,B:\mathcal{H} \to \mathcal{H}$ be self-adjoint operators, $\ket{f} \in \mathcal{H}$ such that $\braket{f}{f} = 1$, then

\begin{equation*}
    (\Delta_f A)(\Delta_f B) \geq \frac{1}{2}\abs{\expected{[A,B]}_f}
\end{equation*}

**Proof:**

Let us define two new operators as

\begin{equation*}
    \begin{split}
        H = A - \expected{A}_f I \\
        G = B - \expected{B}_f I
    \end{split}
\end{equation*}

We can first note that since $A$ and $B$ are self-adjoint operators, $H$ and $G$ are also self-adjoint operators because

\begin{equation*}
    \begin{split}
        H^\dagger &= (A - \expected{A}_f I)^\dagger = A^\dagger - \expected{A}_f I^\dagger = A - \expected{A}_f I = H \\
        G^\dagger &= (B - \expected{B}_f I)^\dagger = B^\dagger - \expected{B}_f I^\dagger = B - \expected{B}_f I = G
    \end{split}
\end{equation*}

Now, in general any expected value and in particular $\expected{HG}_f \in \mathbb{C}$ can be expressed in the following way

\begin{equation*}
    \expected{HG}_f = a + ib
\end{equation*}

Thus

\begin{equation*}
    \expected{GH}_f = \expected{G^\dagger H^\dagger}_f = \expected{(HG)^\dagger}_f = \expected{HG}_f^* = a -ib
\end{equation*}

since $G,H$ are self-adjoint operators. On the other hand

\begin{equation*}
    \begin{split}
        \expected{[H,G]}_f &= \expected{HG-GH}_f = \expected{HG}_f - \expected{GH}_f = (a + ib) - (a - ib) = 2ib \\
        \expected{\left\{H,G\right\}}_f &= \expected{HG+GH}_f = \expected{HG}_f + \expected{GH}_f = (a + ib) + (a - ib) = 2a \\
    \end{split}
\end{equation*}

Let us recall that the expected value is defined in terms of an inner product and therefore we can open up the sum and subtraction in the previous expressions. On the other hand, since a non-negative number is always less than or equal to itself plus another non-negative number, we can write this inequality

\begin{equation*}
    \abs{\expected{[H,G]}_f}^2 \leq \abs{\expected{[H,G]}_f}^2 + \abs{\expected{\left\{H,G\right\}}_f}^2
\end{equation*}

So, we have that

\begin{equation}\label{paso1}
    \begin{split}
        \abs{\expected{[H,G]}_f}^2 &\leq \abs{\expected{[H,G]}_f}^2 + \abs{\expected{\left\{H,G\right\}}_f}^2 \\
        &= \abs{2ib}^2 + \abs{2a}^2 \\
        &= 4b^2 + 4a^2 \\
        &= 4(a^2 + b^2) \\
        &= 4\abs{a + ib}^2 \\
        &= 4 \abs{\expected{HG}_f}^2
    \end{split}
\end{equation}

Considering now

\begin{equation*}
    \begin{split}
        \ket{h} &= H\ket{f} \\
        \ket{g} &= G\ket{f}
    \end{split}
\end{equation*}

Then, by the Cauchy-Schwarz inequality we have that

\begin{equation}\label{paso2}
    \begin{split}
        \abs{\braket{h}{g}}^2 &\leq \braket{h}{h} \braket{g}{g} \implies \\
        \abs{\bra{f}H^\dagger G\ket{f}}^2 &\leq \bra{f}H^\dagger H\ket{f} \bra{f}G^\dagger G\ket{f}
        \implies \\
        \abs{\bra{f}HG\ket{f}}^2 &\leq \bra{f}HH\ket{f} \bra{f}GG\ket{f}
        \implies \\
        \abs{\expected{HG}_f}^2 &\leq \expected{H^2}_f \expected{G^2}_f
    \end{split}
\end{equation}

Now, from these two inequalities we have that

\begin{equation*}
    \abs{\expected{[H,G]}_f}^2 \leq 4 \abs{\expected{HG}_f}^2 \leq 4 \expected{H^2}_f \expected{G^2}_f
\end{equation*}

And from <a href="#proposition_2_25">Proposition 2.25</a> we have that

\begin{equation*}
    \begin{split}
        \expected{[H,G]}_f &= \expected{[A,B]}_f \\
        \expected{H^2}_f &= (\Delta_f A)^2 \\
        \expected{G^2}_f &= (\Delta_f B)^2
    \end{split}
\end{equation*}

Thus

\begin{equation*}
    \abs{\expected{[A,B]}_f}^2 \leq 4 (\Delta_f A)^2 (\Delta_f B)^2
\end{equation*}

Therefore

\begin{equation*}
    (\Delta_f A)(\Delta_f B) \geq \frac{1}{2}\abs{\expected{[A,B]}_f}
\end{equation*}

As we discussed earlier, Heisenberg's Uncertainty Principle is obtained from this result when the operators $A$ and $B$ are the position and momentum operators in a quantum system.

### Example: Heisenberg's Uncertainty Principle

In this example we are going momentarily leave from a finite dimensional space usual in Quantum Computing to an infinite dimensional space and we will see the operators as differential operators.
Let $Q=qI$ be the position operator and $P = -i\hbar\frac{\partial}{\partial q}$ the momentum operator, where $\hbar$ is Planck's Reduced Constant. (Note: These are infinite dimensional operators)

Thus

\begin{equation*}
    \begin{split}
        [Q,P]\ket{\psi(q)} &= (QP-PQ)\ket{\psi(q)} \\
        &= qI(-i\hbar\frac{\partial}{\partial q})\ket{\psi(q)} - (-i\hbar\frac{\partial}{\partial q})qI\ket{\psi(q)} \\
        &= -i\hbar q\frac{\partial}{\partial q}\ket{\psi(q)} + i\hbar\frac{\partial}{\partial q} \left(q \ket{\psi(q)}\right) \\
    \end{split}
\end{equation*}

applying the rule of the derivative of a product to the term on the right, we have

\begin{equation*}
    \begin{split}
        [Q,P]\ket{\psi(q)} &= -i\hbar q\frac{\partial}{\partial q}\ket{\psi(q)} + i\hbar\ket{\psi(q)} + i\hbar q\frac{\partial}{\partial q}\ket{\psi(q)} \\
        &= i\hbar\ket{\psi(q)}
    \end{split}
\end{equation*}

Hence $[Q,P] = i\hbar I$

From <a href="#theorem_2_27">Heisenberg Principle</a> we know that

\begin{equation*}
    \begin{split}
        (\Delta_\psi Q) (\Delta_\psi P) &\geq \frac{1}{2}\abs{\expected{[Q,P]}_f} \\
        &= \frac{1}{2}\abs{\expected{i\hbar I}_f} \\
        &= \frac{1}{2}\abs{i\hbar} \\
        &= \frac{\hbar}{2}
    \end{split}
\end{equation*}

Therefore $(\Delta_\psi Q) (\Delta_\psi P) \geq \frac{\hbar}{2}$
