# Time Evolution Operator

_prepared by Israel Gelover_

The following definition is closely related to <a href="./Postulates.ipynb#definition_3_1">Postulate VII</a> on temporal evolution.

### <a name="definition_3_2">Definition 3.2</a>

The _Time Evolution Operator_ $\hat{T}(t,t_0)$ is defined as:

\begin{equation*}
    \ket{\psi(t)} = \hat{T}(t,t_0)\ket{\psi(t_0)}
\end{equation*}

Note that postulate VII already tells us what the time evolution of a state is like, what we are looking for with this definition is to have a version defined in terms of a linear operator. That is, we want that when applying this operator to the state of the system at time $t_0$ it tells us what the state of the system will be at time $t$ in the future. Up to this point we have not said what this operator is, but we are describing how it should act in the event that
exist.

### <a name="proposition_3_3">Proposition 3.3</a>

By <a href="./Postulates.ipynb#definition_3_1">Postulate I</a>, we know that states are unit vectors, that is

\begin{equation*}
    1 = \braket{\psi(t_0)}{\psi(t_0)} = \braket{\psi(t)}{\psi(t)}
\end{equation*}

From the definition we have that

\begin{equation*}
    \ket{\psi(t)} = \hat{T}(t,t_0)\ket{\psi(t_0)}
\end{equation*}

Thus, rewriting $\ket{\psi(t)}$ we get

\begin{equation*}
    1 = \braket{\psi(t)}{\psi(t)} = \bra{\psi(t_0)}\hat{T}^\dagger(t,t_0) \hat{T}(t,t_0)\ket{\psi(t_0)}
\end{equation*}

That is, the operator $\hat{T}(t,t_0)$ preserves the inner product, therefore $\hat{T}(t,t_0)$ is unitary.

Let's see now that there is indeed a time evolution operator.

### <a name="remark_3_4">Remark 3.4</a>

If the Hamiltonian $\hat{H}$ does not explicitly depend on time $t$ (that is, the system is closed, meaning the energy does not change in time), then the time evolution operator is given by

\begin{equation*}
    \hat{T}(t,t_0) = e^{-\frac{i}{\hbar}(t-t_0)\hat{H}}
\end{equation*}

where the exponentiation of a matrix is defined as $e^{\hat{A}} = \sum_{n=0}^\infty \frac{\hat{A}^{n}}{n!}$, which again is a matrix.

Why? Let's elaborate a bit on the motivation for this claim. Let us consider the following differential equation: A constant $K$ multiplying the derivative of a real function $f$, equal to another constant that we will call $H$, multiplied by the same function $f$.

\begin{equation}\label{edif}
    \begin{split}
        K\dv{f}{t} &= Hf(t) \enspace \text{with } K,H \in \mathbb{R} \\
        \implies \dv{f}{t} &= \frac{H}{K}f(t) \\
        \implies f(t) &= e^{\frac{H}{K}(t-t_0)}f(t_0)
    \end{split}
\end{equation}

where the initial condition of the system is $f(t_0)$.

As an observation that we are going to need, we mention that a result of the Spectral Theory is that self-adjoint matrices are diagonalizable. On the other hand, if we consider Schr√∂dinger Equation

\begin{equation}\label{inicial}
    i\hbar\dv{}{t}\ket{\psi(t)} = \hat{H}\ket{\psi(t)}
\end{equation}

where $\hat{H}$ is a diagonal matrix, what we obtain is a system of equations of type $K\dv{f}{t} = Hf(t)$.

How can we visualize this? Take as an example

\begin{equation*}
    \ket{\psi(t)} = \begin{pmatrix}\psi_1(t) \\ \psi_2(t)\end{pmatrix} \enspace \text{ and } \hat{H} = \begin{pmatrix}h_{11} & 0 \\ 0 & h_{22}\end{pmatrix}
\end{equation*}

in this case we have the following

\begin{equation*}
    \begin{split}
        i\hbar\dv{}{t}\ket{\psi(t)} &= \hat{H}\ket{\psi(t)} \\
        \implies i\hbar\dv{}{t}\begin{pmatrix}\psi_1(t) \\ \psi_2(t)\end{pmatrix} &= \begin{pmatrix}h_{11} & 0 \\ 0 & h_{22}\end{pmatrix}\begin{pmatrix}\psi_1(t) \\ \psi_2(t)\end{pmatrix} \\
        \implies \hbar\dv{}{t}\begin{pmatrix}\psi_1(t) \\ \psi_2(t)\end{pmatrix} &= \begin{pmatrix}h_{11}\psi_1(t) \\ h_{22}\psi_2(t)\end{pmatrix}
    \end{split}
\end{equation*}

that is, we have a system of equations $i\hbar\dv{}{t}\psi_j(t) = h_j\psi_j(t)$ with $j \in \set{1,2}$ that are clearly of the type $K\dv{f}{t} = Hf(t)$. Thus

\begin{equation*}
    \begin{split}
        i\hbar\dv{}{t}\ket{\psi(t)} &= \hat{H}\ket{\psi(t)} \\
        \implies \dv{}{t}\ket{\psi(t)} &= \frac{-i}{\hbar}\hat{H}\ket{\psi(t)} \\
        \implies \ket{\psi(t)} &= e^{-\frac{i}{\hbar}(t-t_0)\hat{H}}\ket{\psi(t_0)}
    \end{split}
\end{equation*}

where the initial condition is the vector $\ket{\psi(t_0)}$.

Being $\hat{H}$ diagonal, the exponential of this matrix is again a diagonal matrix whose entries on the diagonal are the exponential of each of the entries on the diagonal of $\hat{H}$. We can see this in the following way. Let

\begin{equation*}
    \hat{H} = \begin{pmatrix}
        h_{11} & 0 & \cdots & 0 \\ 0 & h_{22} & \cdots & 0 \\ \vdots & \vdots & \cdots & \vdots \\ 0 & 0 & \cdots & h_{mm}
    \end{pmatrix}
    \implies
    \hat{H}^n = \begin{pmatrix}
        h_{11}^n & 0 & \cdots & 0 \\ 0 & h_{22}^n & \cdots & 0 \\ \vdots & \vdots & \cdots & \vdots \\ 0 & 0 & \cdots & h_{mm}^n
    \end{pmatrix}
\end{equation*}

Let's calculate $e^{\hat{H}}$ according to the definition

\begin{equation*}
    \begin{split}
        e^{\hat{H}} &= \sum_{n=0}^\infty \frac{\hat{H}^{n}}{n!} \\
        &= \sum_{n=0}^\infty \frac{1}{n!} \begin{pmatrix}
            h_{11}^n & 0 & \cdots & 0 \\ 0 & h_{22}^n & \cdots & 0 \\ \vdots & \vdots & \cdots & \vdots \\ 0 & 0 & \cdots & h_{mm}^n
        \end{pmatrix} \\
        &= \begin{pmatrix}
            \sum_{n=0}^\infty \frac{h_{11}^n}{n!} & 0 & \cdots & 0 \\ 0 & \sum_{n=0}^\infty \frac{h_{22}^n}{n!} & \cdots & 0 \\ \vdots & \vdots & \cdots & \vdots \\ 0 & 0 & \cdots & \sum_{n=0}^\infty \frac{h_{mm}^n}{n!}
        \end{pmatrix}
    \end{split}
\end{equation*}

But let us recall that $\sum_{n=0}^\infty \frac{z^n}{n!}$ is the expression in _Taylor series_ of $e^{z}$, then we have

\begin{equation*}
    e^{\hat{H}} = \begin{pmatrix}
        e^{h_{11}} & 0 & \cdots & 0 \\ 0 & e^{h_{22}} & \cdots & 0 \\ \vdots & \vdots & \cdots & \vdots \\ 0 & 0 & \cdots & e^{h_{mm}}
    \end{pmatrix}
\end{equation*}

In summary, we can conclude that

\begin{equation}\label{temporal2}
    \ket{\psi(t)} = e^{-\frac{i}{\hbar}(t-t_0)\hat{H}}\ket{\psi(t_0)}
\end{equation}

and therefore the operator

\begin{equation*}
    \hat{T}(t,t_0) = e^{-\frac{i}{\hbar}(t-t_0)\hat{H}}
\end{equation*}

is the time evolution operator we are looking for.

### <a name="proposition_3_5">Proposition 3.5</a>

Let $\hat{H}\ket{E} = E\ket{E}$, where $\ket{E}$ is an eigenvector with eigenvalue $E$, and $f$ an analytic function. Then

\begin{equation*}
    f(\hat{H})\ket{E} = f(E)\ket{E}
\end{equation*}

**Proof:**

Since $f$ is analytical we can express it in terms of its Taylor series, and for simplicity we are going to suppose that the expansion can be done around $0$, that is

\begin{equation*}
    \begin{split}
        f(\hat{H}) &= \sum_{j=0}^\infty \frac{f^{(n)}(0)}{j!}\hat{H}^n \\
        \implies f(\hat{H})\ket{E} &= \sum_{j=0}^\infty \frac{f^{(n)}(0)}{j!}\hat{H}^n\ket{E} \\
        &= \sum_{j=0}^\infty \frac{f^{(n)}(0)}{j!}E^n\ket{E} = f(E)\ket{E}
    \end{split}
\end{equation*}

This result will be particularly useful when we take $f$ as the exponential. Since, if we want to apply the exponential of an operator to any ket, we can write this vector in terms of the eigenvectors of the operator that we have in the exponential and in this way, it is enough to apply the exponential to the eigenvalues of the operator.
