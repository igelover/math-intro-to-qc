# Bell's Inequality

_prepared by Israel Gelover_

John Bell was a physicist who worked at CERN and who agreed with the ideas of EPR, in fact he proposed an experiment that would give the possibility of demonstrating that what was wrong in the EPR paradox was precisely quantum mechanics.

There are several variants of Bell's inequality and the one we will review here is the simplest. It assumes that the results of an experiment are given only by classical correlations, however quantum mechanics shows that these correlations do not agree with what was observed experimentally in the laboratory.

### <a name="definition_4_4">Definition 4.4</a> Bell's Inequality

Let's go back to the idea addressed in the development of the EPR paradox, where we have two particles in the state $\ket{\psi^-}$ with a certain spin; remember that spin is an intrinsic property of particles that can be measured in an arbitrary direction of our three-dimensional space, and something important to note is that regardless of the direction in which it is measured, its eigenvalues are always $+1 $ or $-1$.
    
In this approach to Bell's inequality, we start by making a measurement of the spin in three arbitrary directions $a,b,c \in \mathbb{R}^3$. In this case, we can measure the spin in the direction of the $Z$ axis, in which case the observable would be $\hat{\sigma_z}$, analogously it could be measured in the direction of the $X$ or $Y$ axes, in in which case the observables would be $\hat{\sigma_x}$ and $\hat{\sigma_y}$ respectively. But note that you can actually measure the spin in any arbitrary direction of $\mathbb{R}^3$, then consider the arbitrary directions $a,b$ and $c$.

In this case, we are posing the same scenario as in the EPR paradox, a pair of particles are thrown towards Alice and Bob and each can measure the spin in any of the three directions $a,b$ or $c$. Let us then make a table with all the possibilities of measuring the spin that these observers have, which would be precisely the way to derive the classical correlation that should exist between the results obtained from said measurements.
    
As we already saw in the case of the EPR paradox, the results are anti-correlated, that is, when Alice measures positive, Bob measures negative, as long as they measure in the same direction. In the same thought experiment of the EPR paradox, if we have three possible directions in which each observer can measure the spin of his particle, these $2^3=8$ would be all possible results that can be obtained.

Suppose that $N$ measurements are made in total, where $N = \sum_{i=1}^8 N_i$ con $N_i \geq 0 \enspace \forall i$. That is, $N_i$ is the number of times that the result of the row $i$ was obtained, within the total of $N$ repetitions of the experiment. Thus, we propose the following inequality, which is simply adding a non-negative number to another non-negative one.

\begin{equation*}
    \begin{split}
        N_3 + N_4 &\leq N_3 + N_4 + N_2 + N_7 \\
        \implies \frac{N_3 + N_4}{N} &\leq \frac{N_2 + N_4}{N} + \frac{N_3 + N_7}{N} \\
    \end{split}
\end{equation*}

From a probability point of view, if we know that each of these results was obtained $N_i$ times, let us note that what we are obtaining on the left side of the inequality is the probability that Alice would have measured positive in the direction $a$ and simultaneously Bob would have measured positive but in the direction $b$. That is, the number of times that the event $(N_3 + N_4)$ occurred, divided by the total number of events $(N)$. In the same way $\frac{N_2 + N_4}{N}$, gives us the probability that Alice measures positive in the direction $a$ and simultaneously Bob measures positive in the direction $c$; and $\frac{N_3 + N_7}{N}$ gives us the probability that Alice measures positive in direction $c$ and simultaneously Bob measures positive in direction $b$. That is, what we are getting is

\begin{equation*}
    P(a^+,b^+) \leq P(a^+,c^+) + P(c^+,b^+)
\end{equation*}

Which is in essence the first inequality that Bell developed in his attempt to agree with EPR, what he sought was to demonstrate that if the EPR paradox (or any other experiment in quantum mechanics) satisfies this inequality, then his results are explainable in classical correlation terms, simply by looking at all the possible combinations that can be obtained.

The result that he later arrived at is that this inequality can be violated if entangled quantum states are taken into account.

### <a name="remark_4_5">Remark 4.5</a>

**1.** Given a normalized direction $a \in \mathbb{R}^3$, the observable (self-adjoint operator) associated with measuring the spin of a particle in the direction $a$ is

\begin{equation*}
    \hat{\sigma_a} = a_x\hat{\sigma_x} + a_y\hat{\sigma_y} + a_x\hat{\sigma_z} \enspace \text{ with } a = (a_x,a_y,a_z) \in \mathbb{R}^3
\end{equation*}

Note that measuring the spin in the direction of the $Z$ axis is equivalent to having $a_x = a_y = 0$ and $a_z = 1$ and thus the associated observable is $\hat{\sigma_z}$. Similarly, for a measurement in the direction of the $X$ and $Y$ axes, the associated observable would be $\hat{\sigma_x}$ and $\hat{\sigma_y}$ respectively.

This observable $\hat{\sigma_a}$, like $\hat{\sigma_x}$, $\hat{\sigma_y}$ and $\hat{\sigma_z}$ also has two eigenvalues $\lambda = \pm 1$, that is, in whatever direction the spin is measured, only a result of $+1$ or $-1$ can be obtained. And its eigenvectors, written in terms of the computational basis, that is, the eigenvectors of $\hat{\sigma_z}$, are

\begin{equation*}
    \begin{split}
        \ket{a^+} &= cos\left(\frac{\theta_a}{2}\right)\ket{0} + e^{i\phi_a}\sin\left(\frac{\theta_a}{2}\right)\ket{1} \\
        \ket{a^-} &= e^{-i\phi_a}sen\left(\frac{\theta_a}{2}\right)\ket{0} - \cos\left(\frac{\theta_a}{2}\right)\ket{1}
    \end{split}
\end{equation*}

where $\theta_a$ and $\phi_a$ are the angles of the direction $a$ in spherical coordinates.

**2.** Given two directions $a,b \in \mathbb{R}^3$, the eigenvectors of $\hat{\sigma_a}$ written in terms of the base eigenvectors of $\hat{\sigma_b}$ are:

\begin{equation*}
    \begin{split}
        \ket{a^+} &= cos\left(\frac{\theta_{ab}}{2}\right)\ket{b^+} + e^{i\phi_{ab}}\sin\left(\frac{\theta_{ab}}{2}\right)\ket{b^-} \\
        \ket{a^-} &= e^{-i\phi_{ab}}sen\left(\frac{\theta_{ab}}{2}\right)\ket{b^+} - \cos\left(\frac{\theta_{ab}}{2}\right)\ket{b^-}
    \end{split}
\end{equation*}

where $\theta_{ab} = \theta_a - \theta_b$, $\phi_{ab} = \phi_a - \phi_b$.

**3.** The entangled state $\ket{\psi^-}$ is written in the same way in any of the bases of the observables of measuring the spin in a direction $a$.

\begin{equation*}
    \ket{\psi^-} = \frac{1}{\sqrt{2}}(\ket{a^+a^-} - \ket{a^-a^+}) \enspace \forall a \in \mathbb{R}^3
\end{equation*}

Where $\ket{a^+}$ is the eigenvector associated with the positive eigenvalue, and $\ket{a^-}$ is the eigenvector associated with the negative eigenvalue. That is, regardless of the direction $a$ we take, when we consider the eigenvectors of $\hat{\sigma_a}$ as a basis, the state $\ket{\psi^-}$ is written in the same way but in terms of the base of eigenvectors of $\hat{\sigma_a}$.

Note that for the observable $\hat{\sigma_z}$ we simply have what we already knew

\begin{equation*}
    \ket{\psi^-} = \frac{1}{\sqrt{2}}(\ket{01} - \ket{10})
\end{equation*}

### <a name="theorem_4_6">Theorem 4.6</a> Bell Theorem

Quantum mechanics does not satisfy Bell's inequality.

**Proof:**

Let us consider that the state of the system is given by the state $\ket{\psi^-}$, and let's calculate the probabilities $P(a^+, b^+), P(a^+, c^+)$ and $P(c^+, b^+)$ in a quantum way, that is:

\begin{equation*}
    \begin{split}
        P(a^+, b^+) &= \abs{\braket{a^+ b^+}{\psi^-}}^2 \\
        P(a^+, c^+) &= \abs{\braket{a^+ c^+}{\psi^-}}^2 \\
        P(c^+, b^+) &= \abs{\braket{c^+ b^+}{\psi^-}}^2
    \end{split}
\end{equation*}

By section **1.** of the previous Remark, we can express $\ket{a^+}$ in terms of the eigenvectors of the base of $\hat{\sigma_b}$ as

\begin{equation*}
    \ket{a^+} = cos\left(\frac{\theta_{ab}}{2}\right)\ket{b^+} + e^{i\phi_{ab}}\sin\left(\frac{\theta_{ab}}{2}\right)\ket{b^-}
\end{equation*}

where $\theta_{ab} = \theta_a - \theta_b$, $\phi_{ab} = \phi_a - \phi_b$.

Furthermore, by section **3.** of the same Remark, we can express $\ket{\psi^-}$ in terms of the same base of $\hat{\sigma_b}$ as

\begin{equation*}
    \ket{\psi^-} = \frac{1}{\sqrt{2}}(\ket{b^+b^-} - \ket{b^-b^+})
\end{equation*}

In this way, let's calculate

\begin{equation*}
    \begin{split}
        P(a^+, b^+) &= \abs{\braket{a^+ b^+}{\psi^-}}^2 \\
        &= \abs{\left[cos\left(\frac{\theta_{ab}}{2}\right)\bra{b^+b^+} + e^{-i\phi_{ab}}\sin\left(\frac{\theta_{ab}}{2}\right)\bra{b^-b^+}\right] \left[\frac{1}{\sqrt{2}}(\ket{b^+b^-} - \ket{b^-b^+})\right]}^2 \\
        &= \abs{\frac{1}{\sqrt{2}} e^{-i\phi_{ab}}\sin\left(\frac{\theta_{ab}}{2}\right)}^2 = \frac{1}{2}\sin^2\left(\frac{\theta_{ab}}{2}\right)
    \end{split}
\end{equation*}

In the same way we can express $\ket{a^+}$ and $\ket{\psi^-}$ in terms of the eigenvectors of the base of $\hat{\sigma_c}$ and calculate

\begin{equation*}
    \begin{split}
        P(a^+, c^+) &= \abs{\braket{a^+ c^+}{\psi^-}}^2 \\
        &= \abs{\left[cos\left(\frac{\theta_{ac}}{2}\right)\bra{c^+c^+} + e^{-i\phi_{ac}}\sin\left(\frac{\theta_{ac}}{2}\right)\bra{c^-c^+}\right] \left[\frac{1}{\sqrt{2}}(\ket{c^+c^-} - \ket{c^-c^+})\right]}^2 \\
        &= \abs{\frac{1}{\sqrt{2}} e^{-i\phi_{ac}}\sin\left(\frac{\theta_{ac}}{2}\right)}^2 = \frac{1}{2}\sin^2\left(\frac{\theta_{ac}}{2}\right)
    \end{split}
\end{equation*}

where $\theta_{ac} = \theta_a - \theta_c$, $\phi_{ac} = \phi_a - \phi_c$.

And finally, we express $\ket{c^+}$ in terms of the base eigenvectors of $\hat{\sigma_b}$ and calculate

\begin{equation*}
    \begin{split}
        P(c^+, b^+) &= \abs{\braket{c^+ b^+}{\psi^-}}^2 \\
        &= \abs{\left[cos\left(\frac{\theta_{cb}}{2}\right)\bra{b^+b^+} + e^{-i\phi_{cb}}\sin\left(\frac{\theta_{cb}}{2}\right)\bra{b^-b^+}\right] \left[\frac{1}{\sqrt{2}}(\ket{b^+b^-} - \ket{b^-b^+})\right]}^2 \\
        &= \abs{\frac{1}{\sqrt{2}} e^{-i\phi_{cb}}\sin\left(\frac{\theta_{cb}}{2}\right)}^2 = \frac{1}{2}\sin^2\left(\frac{\theta_{cb}}{2}\right)
    \end{split}
\end{equation*}

where $\theta_{cb} = \theta_c - \theta_b$, $\phi_{cb} = \phi_a - \phi_b$.

Now, let's consider that the directions $a,b$ and $c$ look as follows

<img src="../img/bell.png" width="220"/>

that is, $\theta_{ab} = 2\theta, \theta_{ac} = \theta$ y $\theta_{cb} = \theta$.

In this way, as

\begin{equation*}
    \theta_{ab} = 2\theta, \mbox{ y } \theta_{ac} = \theta = \theta_{cb}
\end{equation*}

we can rewrite the probabilities as follows

\begin{equation*}
    \begin{split}
        P(a^+, b^+) &= \frac{1}{2}\sin^2\left(\frac{\theta_{ab}}{2}\right) = \frac{1}{2}\sin^2(\theta) \\
        P(a^+, c^+) &= \frac{1}{2}\sin^2\left(\frac{\theta_{ac}}{2}\right) = \frac{1}{2}\sin^2\left(\frac{\theta}{2}\right) \\
        P(c^+, b^+) &= \frac{1}{2}\sin^2\left(\frac{\theta_{cb}}{2}\right) = \frac{1}{2}\sin^2\left(\frac{\theta}{2}\right)
    \end{split}
\end{equation*}

Thus, to obtain the values of $\theta$ for which

\begin{equation*}
    P(a^+, b^+) > P(a^+, c^+) + P(c^+, b^+)
\end{equation*}

we need to solve the following inequality for $\theta$

\begin{equation*}
    \begin{split}
        \frac{1}{2}\sin^2(\theta) &> \frac{1}{2}\sin^2\left(\frac{\theta}{2}\right) + \frac{1}{2}\sin^2\left(\frac{\theta}{2}\right) \\
        \implies \frac{1}{2}\sin^2(\theta) &> \sin^2\left(\frac{\theta}{2}\right) \\
        \implies \sin^2(\theta) &> 2\sin^2\left(\frac{\theta}{2}\right)
    \end{split}
\end{equation*}

which has real solutions for

\begin{equation*}
    2\pi n - \frac{\pi}{2} < \theta < 2\pi n \enspace \mbox{ y } \enspace 2\pi n < \theta < 2\pi n + \frac{\pi}{2}, \enspace n \in \mathbb{Z}
\end{equation*}

Therefore, Quantum Mechanics does not satisfy Bell's inequality.

Note that the philosophical and physical implications of this result are profound. The underlying reason, at least the most widely accepted, why quantum mechanics does not satisfy this inequality, is because when constructing the previous table, it is assumed that the particle that the observer measures has a state and that at the time of being measured, you are simply obtaining information from that state that the particle already had. But quantum mechanics does not satisfy this, since by <a href="./Entanglement.ipynb#definition_4_2">Definition 4.2</a>, when we have a composite system, there are certain (entangled) states, which cannot be factored by independent states, that is, it cannot be thought that each particle has its own state, but the only way to describe the state of both is together and when making a measurement on one of them, automatically the state of the other also changes, and changes instantaneously.
