### <a name="definition_1_3">Definition 1.3</a> Bit
A bit is a variable that can take one of two possible values and is the minimum unit of classical information. 
$$ b \in \set{0,1} $$

#### Example:
- In an electrical circuit, we can have a terminal at 0V (bit 0) or at 5V (bit 1).
- A LED can be off (bit 0) or on (bit 1).

## Logic Gates

Bit manipulatation happens through a logic gate paradigm.

### <a name="definition_1_4">Definition 1.4</a> Logic Gate
A logic gate is a function
$$ f:\set{0,1}^n \to \set{0,1}^m \enspace n,m \in \mathbb{N} $$

There are several ways to get to the paradigm of classical computing, one of the best known is through Turing machines. The set of computable functions (operations that can be done) with the formalization of Turing machines turns out to be the same set of computable functions that is obtained through the logic gates paradigm.

#### Example:

- Identity
\begin{equation*}
    \begin{split}
        \mbox{I}:\set{0,1} &\to \set{0,1} \\
        a &\to a
    \end{split}
\end{equation*}
- Not
\begin{equation*}
    \begin{split}
        \mbox{Not}:\set{0,1} &\to \set{0,1} \\
        a &\to \overline{a} = 1-a
    \end{split}
\end{equation*}
- And
\begin{equation*}
    \begin{split}
        \land:\set{0,1}^2 &\to \set{0,1} \\
        (a,b) &\to a \land b = a \cdot b
    \end{split}
\end{equation*}
- Or
\begin{equation*}
    \begin{split}
        \lor:\set{0,1}^2 &\to \set{0,1} \\
        (a,b) &\to a \lor b = a + b - a \cdot b
    \end{split}
\end{equation*}
- XOr
\begin{equation*}
    \begin{split}
        \oplus:\set{0,1}^2 &\to \set{0,1} \\
        (a,b) &\to a \oplus b = a + b \mod{2} = a + b - 2ab
    \end{split}
\end{equation*}
- NAnd = Not $\circ$ And
- NOr = Not $\circ$ Or
- Copy
\begin{equation*}
    \begin{split}
        \mbox{Copy}:\set{0,1} &\to \set{0,1}^2 \\
        a &\to (a,a)
    \end{split}
\end{equation*}
- Swap
\begin{equation*}
    \begin{split}
        \mbox{Swap}:\set{0,1}^2 &\to \set{0,1}^2 \\
        (a,b) &\to (b,a)
    \end{split}
\end{equation*}
- Toffoli
\begin{equation*}
    \begin{split}
        \mbox{Toffoli}:\set{0,1}^3 \rightarrow \set{0,1}^3 \\\\
        \mbox{Toffoli}(0,0,0) = (0,0,0), \\
        \mbox{Toffoli}(0,0,1) = (0,0,1), \\
        \mbox{Toffoli}(0,1,0) = (0,1,0), \\
        \mbox{Toffoli}(0,1,1) = (0,1,1), \\
        \mbox{Toffoli}(1,0,0) = (1,0,0), \\
        \mbox{Toffoli}(1,0,1) = (1,0,1), \\
        \mbox{Toffoli}(1,1,0) = (1,1,1), \\
        \mbox{Toffoli}(1,1,1) = (1,1,0). \\        
    \end{split}
\end{equation*}
where $\circ$ denotes the usual function composition.

### <a name="theorem_1_5">Theorem 1.5</a> De Morgan's Laws

1. $\overline{a \land b} = \overline{a} \lor \overline{b}$  
    $Not(And(a,b)) = Or(Not(a), Not(b))$
    
2. $\overline{a \lor b} = \overline{a} \land \overline{b}$  
    $Not(Or(a,b)) = And((Not(a), Not(b)))$

_Proof:_
From the definition of Not, And y Or, we have  
1.
\begin{equation*}
    \begin{split}
        \overline{a} \lor \overline{b} &= \overline{a} + \overline{b} - (\overline{a} \cdot \overline{b}) \\
        &= (1 - a) + (1 - b) - (1 - a)(1 - b) \\
        &= 1 - a + 1 - b - (1 - b - a + (a \cdot b)) \\
        &= 1 - a + 1 - b - 1 + b + a - (a \cdot b) \\
        &= 1 - (a \cdot b) \\
        &= \overline{a \land b}
    \end{split}
\end{equation*}
2.
\begin{equation*}
    \begin{split}
        \overline{a \lor b} &= 1 - (a + b - (a \cdot b)) \\
        &= 1 - a - b + (a \cdot b) \\
        &= (1 - a) - b(1-a) \\
        &= (1 - a)(1 - b) \\
        &= \overline{a} \land \overline{b}
    \end{split}
\end{equation*}

### <a name="theorem_1_7">Theorem 1.7</a> Conjunctive normal form
Any logic gate $f:\set{0,1}^n \to \set{0,1}^m$ can be built using gates And and Not.

_Proof:_ Let $f:\set{0,1}^n \to \set{0,1}^m$ be a logic gate.  
We can express $f = (f_1, f_2, \dots, f_m)$ where each $f_j:\set{0,1}^n \to \set{0,1}, j \in \set{1, \dots, m}$

Let $k \in \set{1, \dots, m}$, we can define $S_k = \set{\overline{a} \in \set{0,1}^n \mid f_k(\overline{a}) = 1}$  
Given $\overline{a}^{(l_k)} \in S_k$ with $l_k \in \set{1, \dots, \abs{S_k}}$, let us define

\begin{equation*}
    \begin{split}
        f_k^{(l_k)}:\set{0,1}^n &\to \set{0,1}\\
        \overline{a} &\to b_1 \land b_2 \land \dots \land b_n
    \end{split}
\end{equation*}

Where
$b_j = \begin{cases}
    a_j & \text{if } a_j^{(l_k)} = 1 \\
    \overline{a_j} & \text{if } a_j^{(l_k)} = 0
\end{cases}$

Note that $f_k^{(l_k)}\left(\overline{a}^{(l_k)}\right) = 1$ and $f_k^{(l_k)}\left(\overline{a}\right) = 0 \enspace \forall \overline{a} \neq \overline{a}^{(l_k)}$
<p style='text-align: right;'>(1)</p>
\begin{equation}
    \implies \overbrace{\underbrace{f_k(\overline{a})}_{0 \enspace \overline{a} \not\in S_k}}^{1 \enspace \overline{a} \in S_k} = \overbrace{\underbrace{f_k^{(1)}(\overline{a}) \lor f_k^{(2)}(\overline{a}) \lor \dots \lor f_k^{(\abs{S_k})}(\overline{a})}_{\text{all equal }0}}^{\text{some equal }1}
\end{equation}

Because

\begin{equation*}
    \begin{split}
        \text{if } \overline{a} \in S_k &\implies \exists l_k \text{ such that } f_k^{(l_k)}(\overline{a}) = 1\\
        \text{if } \overline{a} \not\in S_k &\implies \forall l_k, f_k^{(l_k)}(\overline{a}) = 0\\
    \end{split}
\end{equation*}

Equation (1) tells us that we have constructed $f_k$ using Or and $f_k^{(l_k)}$ gates. On the other hand, each gate $f_k^{(l_k)}$ was built using only And and Not gates. And by De Morgan's Laws, Or can be built using only And and Not gates, and vice versa, we can build the And gate using only gates Or and Not.

**Remark:** The set {And, Not} is called a universal set of logic gates. The set {Or, Not} is also universal and is known as the _Disjunctive normal form_.

### <a name="proposition_1_8">Proposition 1.8</a>
