# Single Qubit Gates

_prepared by Israel Gelover_

In this notebook we will give some results with single qubit gates as a motivation to see that to build any gate of a single qubit, we only need the Hadamard gate and $R_z$ rotations, later we will see the main multi-qubit gates, to finally arrive to the main result of this section, the _Solovay-Kitaev Theorem_. This theorem provides a finite set of gates that is computationally universal.

### <a name="remark_5_8">Remark 5.8</a> Single Qubit Gates

**1.** Putting an $R_z$ gate between two Hadamard gates is equivalent to a rotation $R_x$ by the same angle. This is

\begin{equation*}
    H R_z(\theta) H = R_x(\theta)
\end{equation*}

Doing the calculation we have

\begin{equation*}
    \begin{split}
        H R_z(\theta) H &= \frac{1}{\sqrt{2}} \begin{pmatrix}1 & 1 \\ 1 & -1\end{pmatrix} \begin{pmatrix}e^{-i\frac{\theta}{2}} & 0 \\ 0 & e^{i\frac{\theta}{2}}\end{pmatrix} \frac{1}{\sqrt{2}} \begin{pmatrix}1 & 1 \\ 1 & -1\end{pmatrix} \\
        &= \frac{1}{2} \begin{pmatrix}1 & 1 \\ 1 & -1\end{pmatrix} \begin{pmatrix}e^{-i\frac{\theta}{2}} & e^{-i\frac{\theta}{2}} \\ e^{i\frac{\theta}{2}} & e^{i\frac{\theta}{2}}\end{pmatrix} \\
        &= \frac{1}{2} \begin{pmatrix}e^{-i\frac{\theta}{2}} + e^{i\frac{\theta}{2}} & e^{-i\frac{\theta}{2}} -e^{i\frac{\theta}{2}} \\ e^{-i\frac{\theta}{2}} - e^{i\frac{\theta}{2}} & e^{-i\frac{\theta}{2}} + e^{i\frac{\theta}{2}}\end{pmatrix}
    \end{split}
\end{equation*}

From <a href="./Rotation.ipynb#theorem_5_5">Theorem 5.5</a> we know that

\begin{equation*}
    \begin{split}
        e^{-i\frac{\theta}{2}} + e^{i\frac{\theta}{2}} &= \cos\left(\frac{\theta}{2}\right) - i\sin\left(\frac{\theta}{2}\right) + \cos\left(\frac{\theta}{2}\right) + i\sin\left(\frac{\theta}{2}\right) = 2\cos\left(\frac{\theta}{2}\right) \\
        e^{-i\frac{\theta}{2}} - e^{i\frac{\theta}{2}} &= \cos\left(\frac{\theta}{2}\right) - i\sin\left(\frac{\theta}{2}\right) - \cos\left(\frac{\theta}{2}\right) - i\sin\left(\frac{\theta}{2}\right) = -2i\sin\left(\frac{\theta}{2}\right)
    \end{split}
\end{equation*}

Therefore

\begin{equation*}
    H R_z(\theta) H = \frac{1}{2} \begin{pmatrix}
        2\cos\left(\frac{\theta}{2}\right) & -2i\sin\left(\frac{\theta}{2}\right) \\ -2i\sin\left(\frac{\theta}{2}\right) & 2\cos\left(\frac{\theta}{2}\right)
    \end{pmatrix} = \begin{pmatrix}
        \cos\left(\frac{\theta}{2}\right) & -i\sin\left(\frac{\theta}{2}\right) \\ -i\sin\left(\frac{\theta}{2}\right) & \cos\left(\frac{\theta}{2}\right)
    \end{pmatrix} = R_x(\theta)
\end{equation*}

**2.** Let

\begin{equation*}
    \begin{split}
        \ket{\psi} &\in \mathbb{B} \text{ with angles } (\theta_1, \phi_1) \\
        \ket{\varphi} &\in \mathbb{B} \text{ with angles } (\theta_2, \phi_2)
    \end{split}
\end{equation*}

The logic gate $U$ that transforms $\ket{\psi}$ into $\ket{\varphi}$, that is, such that $U\ket{\psi} = \ket{\varphi}$ is of the form

\begin{equation*}
    U = R_z\left(\frac{\pi}{2} + \phi_2\right) \underbrace{H R_z(\theta_2 - \theta_1) H}_{R_x(\theta_2 - \theta_1)} R_z\left(-\frac{\pi}{2} - \phi_1\right)
\end{equation*}

Let's see how we can interpret this gate geometrically. First we note that since the qubit $\ket{\psi}$ is parameterized with the angle $\phi_1$, we can say that the rotation $R_z\left(-\frac{\pi}{2} - \phi_1\right)$ is returning the vector to the $XZ$ plane; then we rotate around the $X$ axis by the difference of the angles $\theta_2 - \theta_1$, that is, we remove the component of the angle $\theta_1$ from $\ket{\psi}$ and add the component of the angle $\theta_2$ from $\ket{\varphi}$; and finally with the gate $R_z\left(\frac{\pi}{2} + \phi_2\right)$ we rotate around the $Z$ axis by an angle $\phi_2$. In this way, the resulting vector is parameterized by the angles $\theta_2, \phi_2$, and thus we obtain the expected result.

It is important to mention that this gate is not the only one that gives us the expected result, however it is of particular interest, since only $H$ and $R_z$ gates were needed to perform this transformation. That is, we can express the gate that takes us from any qubit $\ket{\psi}$ to any other qubit $\ket{\varphi}$ using only Hadamard and $R_z$ rotations.

**3.** A result similar to **1.** tells us that

\begin{equation*}
    R_y(\theta) = R_z\left(\frac{\pi}{2}\right)R_x(\theta)R_z\left(-\frac{\pi}{2}\right)
\end{equation*}

Doing the calculations we have

\begin{equation*}
    \begin{split}
        R_z\left(\frac{\pi}{2}\right)R_x(\theta)R_z\left(-\frac{\pi}{2}\right) &= \begin{pmatrix}e^{-i\frac{\pi}{4}} & 0 \\ 0 & e^{i\frac{\pi}{4}}\end{pmatrix} \begin{pmatrix}\cos\left(\frac{\theta}{2}\right) & -i\sin\left(\frac{\theta}{2}\right) \\ -i\sin\left(\frac{\theta}{2}\right) & \cos\left(\frac{\theta}{2}\right)\end{pmatrix} \begin{pmatrix}e^{i\frac{\pi}{4}} & 0 \\ 0 & e^{-i\frac{\pi}{4}}\end{pmatrix} \\
        &= \begin{pmatrix}e^{-i\frac{\pi}{4}} & 0 \\ 0 & e^{i\frac{\pi}{4}}\end{pmatrix} \begin{pmatrix}\cos\left(\frac{\theta}{2}\right)e^{i\frac{\pi}{4}} & -i\sin\left(\frac{\theta}{2}\right)e^{-i\frac{\pi}{4}} \\ -i\sin\left(\frac{\theta}{2}\right)e^{i\frac{\pi}{4}} & \cos\left(\frac{\theta}{2}\right)e^{-i\frac{\pi}{4}}\end{pmatrix} \\
        &= \begin{pmatrix}\cos\left(\frac{\theta}{2}\right) & -i\sin\left(\frac{\theta}{2}\right)e^{-i\frac{\pi}{2}} \\ -i\sin\left(\frac{\theta}{2}\right)e^{i\frac{\pi}{2}} & \cos\left(\frac{\theta}{2}\right)\end{pmatrix}
    \end{split}
\end{equation*}

Since $e^{i\frac{\pi}{2}} = i$ y $e^{-i\frac{\pi}{2}} = -i$, then

\begin{equation*}
    R_z\left(\frac{\pi}{2}\right)R_x(\theta)R_z\left(-\frac{\pi}{2}\right) = \begin{pmatrix}
        \cos\left(\frac{\theta}{2}\right) & -\sin\left(\frac{\theta}{2}\right) \\ \sin\left(\frac{\theta}{2}\right) & \cos\left(\frac{\theta}{2}\right)
    \end{pmatrix} = R_y(\theta)
\end{equation*}

**4.** Let $U:\mathbb{B} \to \mathbb{B}$ be a single qubit quantum gate, then there are $\alpha,\beta,\gamma,\delta \in \mathbb{R}$ such that

\begin{equation*}
    U = R_z(\alpha)R_y(\beta)R_z(\gamma) e^{i\delta}
\end{equation*}

Note that from **3.** $R_y(\beta) = R_z\left(\frac{\pi}{2}\right)R_x(\beta)R_z\left(-\frac{\pi}{2}\right)$, and by **1.**  we can write the gate $R_x$ in terms of a gate $R_z$ between two Hadamard gates. In summary, the conclusion we are reaching is that given any single-qubit gate $U$, we can express it in terms of only $H$ and $R_z$ gates.

**5.** Let $U:\mathbb{B} \to \mathbb{B}$ be a single qubit quantum logic gate, then there exists $V$ single qubit gate such that $V^2 = U$, i.e. $V = \sqrt{U}$

Let $U = \begin{pmatrix}a & b \\ c & d\end{pmatrix} \enspace \mbox{con } a,b,c,d \in \mathbb{C}$. Let us consider

\begin{equation*}
    s^2 = ad-bc \enspace\mbox{ and }\enspace t^2 = a + d + 2s
\end{equation*}

that is

\begin{equation*}
    s = \pm\sqrt{ad-bc} \enspace\mbox{ and }\enspace t = \pm\sqrt{a + d + 2s}
\end{equation*}

If $t \neq 0$, then the following matrix $V$ satisfies that $V^2 = U$

\begin{equation*}
    V = \frac{1}{t} \begin{pmatrix}a+s & b \\ c & d+s\end{pmatrix} = \frac{1}{t}(U + sI)
\end{equation*}

We can verify this statement calculating

\begin{equation*}
    \begin{split}
        V^2 &= \frac{1}{t} \begin{pmatrix}a+s & b \\ c & d+s\end{pmatrix} \frac{1}{t} \begin{pmatrix}a+s & b \\ c & d+s\end{pmatrix} \\
        &= \frac{1}{t^2} \begin{pmatrix}
            a^2 + 2as + s^2 + bc & ab + sb + bd + bs \\ ca + cs + dc + sc & cb + d^2 + 2ds + s^2
        \end{pmatrix} \\
        &= \frac{1}{t^2} \begin{pmatrix}
            a^2 + 2as + ad - bc + bc & ab + bd + 2sb \\ ca + dc + 2sc & cb + d^2 + 2ds + ad - bc
        \end{pmatrix} \\
        &= \frac{1}{t^2} \begin{pmatrix}
            a^2 + 2as + ad & ab + bd + 2sb \\ ca + dc + 2sc & d^2 + 2ds + ad
        \end{pmatrix} \\
        &= \frac{1}{a+d+2s} \begin{pmatrix}
            a(a + 2s + d) & b(a + d + 2s) \\ c(a + d + 2s) & d(d + 2s + a)
        \end{pmatrix} \\
        &= \begin{pmatrix}a & b \\ c & d\end{pmatrix} \\
        &= U
    \end{split}
\end{equation*}
