# Rotation Gates

_prepared by Israel Gelover_

Let us recall from <a href="../2-math/BlochSphere.ipynb#definition_2_8">Definition 2.8</a> that a qubit is essentially a point on the Bloch sphere. The effect that the rotation gates have on a qubit is precisely to rotate this point on the sphere and in this way we obtain a new point (qubit) on the sphere. We can also understand the effect of these gates as a rotation of the Bloch sphere. In this way, there are an infinite number of ways to rotate a sphere and therefore an infinite number of rotation gates, in this section we are going to define only the three main ones, which are rotations with respect to the $X$, $Y$ and $Z$ axes.

### <a name="theorem_5_5">Theorem 5.5</a>

Let $A:\mathcal{H} \to \mathcal{H}$ be such that $A^2 = I$, then

\begin{equation*}
    e^{-i\theta A} = \cos(\theta)I - i\sin(\theta)A
\end{equation*}

**Proof:**

Recall that by definition, we can see the exponential of an operator in its Taylor series expansion, like this

\begin{equation*}
    e^{-i\theta A} = \sum_{k=0}^\infty \frac{(-i\theta A)^k}{k!}
\end{equation*}

Let's separate the sum into its odd and even parts. Let us note that particularly if $k$ is even, that is, it is of the form $k=2n$, then $-i^k = -i^{2n} = (-1)^n$, and also, since $A^2 = I$, then $A^k = A^{2n} = I$. On the other hand, if $k$ is odd, that is, it is of the form $k=2n+1$, then $-i^k = -i^{2n+1} = (-i)^{2n}(-i) =(-1)^n(-i)$, and also $A^k = A^{2n+1} = A^{2n}A = A$. Thus

\begin{equation*}
    \begin{split}
        e^{-i\theta A} &= \left[\sum_{n=0}^\infty \frac{(-1)^n \theta^{2n}}{(2n)!}\right]I - i\left[\sum_{n=0}^\infty \frac{(-1)^n \theta^{2n+1}}{(2n+1)!}\right]A \\
        &= \cos(\theta)I - i\sin(\theta)A
    \end{split}
\end{equation*}

Again, using the Taylor series expansion of $\sin(\theta)$ and $\cos(\theta)$.

### <a name="definition_5_6">Definition 5.6</a> Rotation Gates with respect to $X$, $Y$ and $Z$ axes

We define the rotation gates with respect to the $X$, $Y$ and $Z$ axes as

\begin{equation*}
    \begin{split}
        R_z(\delta) &:= e^{-i\frac{\delta}{2}\hat{\sigma_z}} \\
        R_y(\delta) &:= e^{-i\frac{\delta}{2}\hat{\sigma_y}} \\
        R_x(\delta) &:= e^{-i\frac{\delta}{2}\hat{\sigma_x}}
    \end{split}
\end{equation*}

Since $\hat{\sigma_x}^2 = \hat{\sigma_y}^2 = \hat{\sigma_z}^2 = I$, we can use **Theorem 5.6** to write these gates in their matrix form in the following way

\begin{equation*}
    \begin{split}
        R_z(\delta) &:= e^{-i\frac{\delta}{2}\hat{\sigma_z}} \\
        &= \cos\left(\frac{\delta}{2}\right)I - i\sin\left(\frac{\delta}{2}\right)\hat{\sigma_z} \\
        &= \cos\left(\frac{\delta}{2}\right)\begin{pmatrix}1 & 0 \\ 0 & 1\end{pmatrix} - i\sin\left(\frac{\delta}{2}\right)\begin{pmatrix}1 & 0 \\ 0 & -1\end{pmatrix} \\
        &= \begin{pmatrix}
            \cos\left(\frac{\delta}{2}\right) - i\sin\left(\frac{\delta}{2}\right) & 0 \\
            0 & \cos\left(\frac{\delta}{2}\right) + i\sin\left(\frac{\delta}{2}\right)
        \end{pmatrix} \\
        &= \begin{pmatrix}
            e^{-i\frac{\delta}{2}} & 0 \\ 0 & e^{i\frac{\delta}{2}}
        \end{pmatrix}
    \end{split}
\end{equation*}

In the same way for $R_y$:

\begin{equation*}
    \begin{split}
        R_y(\delta) &:= e^{-i\frac{\delta}{2}\hat{\sigma_y}} \\
        &= \cos\left(\frac{\delta}{2}\right)I - i\sin\left(\frac{\delta}{2}\right)\hat{\sigma_y} \\
        &= \cos\left(\frac{\delta}{2}\right)\begin{pmatrix}1 & 0 \\ 0 & 1\end{pmatrix} - i\sin\left(\frac{\delta}{2}\right)\begin{pmatrix}0 & -i \\ i & 0\end{pmatrix} \\
        &= \begin{pmatrix}
            \cos\left(\frac{\delta}{2}\right) & -\sin\left(\frac{\delta}{2}\right) \\ \sin\left(\frac{\delta}{2}\right) & \cos\left(\frac{\delta}{2}\right)
        \end{pmatrix}
    \end{split}
\end{equation*}

And also for $R_x$:

\begin{equation*}
    \begin{split}
        R_x(\delta) &:= e^{-i\frac{\delta}{2}\hat{\sigma_x}} \\
        &= \cos\left(\frac{\delta}{2}\right)I - i\sin\left(\frac{\delta}{2}\right)\hat{\sigma_x} \\
        &= \cos\left(\frac{\delta}{2}\right)\begin{pmatrix}1 & 0 \\ 0 & 1\end{pmatrix} - i\sin\left(\frac{\delta}{2}\right)\begin{pmatrix}0 & 1 \\ 1 & 0\end{pmatrix} \\
        &= \begin{pmatrix}
            \cos\left(\frac{\delta}{2}\right) & - i\sin\left(\frac{\delta}{2}\right) \\
            - i\sin\left(\frac{\delta}{2}\right) & \cos\left(\frac{\delta}{2}\right)
        \end{pmatrix}
    \end{split}
\end{equation*}

In summary, we have

\begin{equation*}
    \begin{split}
        R_z(\delta) &= e^{-i\frac{\delta}{2}\hat{\sigma_z}} = \begin{pmatrix}
            e^{-i\frac{\delta}{2}} & 0 \\ 0 & e^{i\frac{\delta}{2}}
        \end{pmatrix} \\
        R_y(\delta) &= e^{-i\frac{\delta}{2}\hat{\sigma_y}} = \begin{pmatrix}
            \cos\left(\frac{\delta}{2}\right) & -\sin\left(\frac{\delta}{2}\right) \\ \sin\left(\frac{\delta}{2}\right) & \cos\left(\frac{\delta}{2}\right)
        \end{pmatrix} \\
        R_x(\delta) &= e^{-i\frac{\delta}{2}\hat{\sigma_x}} = \begin{pmatrix}
            \cos\left(\frac{\delta}{2}\right) & -i\sin\left(\frac{\delta}{2}\right) \\ -i\sin\left(\frac{\delta}{2}\right) & \cos\left(\frac{\delta}{2}\right)
        \end{pmatrix} \\
    \end{split}
\end{equation*}

Let's see how these gates transform to the elements of the computational base, that is $\ket{0}$ and $\ket{1}$.

\begin{equation*}
    \begin{split}
        R_z(\delta)\ket{0} &= \begin{pmatrix}e^{-i\frac{\delta}{2}} & 0 \\ 0 & e^{i\frac{\delta}{2}}\end{pmatrix} \begin{pmatrix}1 \\ 0\end{pmatrix} = \begin{pmatrix}e^{-i\frac{\delta}{2}} \\ 0\end{pmatrix}  = e^{-i\frac{\delta}{2}} \ket{0} = \ket{0} \\
        R_z(\delta)\ket{1} &= \begin{pmatrix}e^{-i\frac{\delta}{2}} & 0 \\ 0 & e^{i\frac{\delta}{2}}\end{pmatrix} \begin{pmatrix}0 \\ 1\end{pmatrix} = \begin{pmatrix}0 \\ e^{i\frac{\delta}{2}}\end{pmatrix}  = e^{i\frac{\delta}{2}} \ket{1} = \ket{1}
    \end{split}
\end{equation*}

since by <a href="../3-qm/Postulates.ipynb#definition_3_1">Postulate I</a>, we have global phase invariance. On the other hand

\begin{equation*}
    \begin{split}
        R_y(\delta)\ket{0} &= \begin{pmatrix}
            \cos\left(\frac{\delta}{2}\right) & -\sin\left(\frac{\delta}{2}\right) \\ \sin\left(\frac{\delta}{2}\right) & \cos\left(\frac{\delta}{2}\right)
        \end{pmatrix} \begin{pmatrix}
            1 \\ 0
        \end{pmatrix} = \begin{pmatrix}
            \cos\left(\frac{\delta}{2}\right) \\ \sin\left(\frac{\delta}{2}\right)
        \end{pmatrix} = \cos\left(\frac{\delta}{2}\right)\ket{0} + \sin\left(\frac{\delta}{2}\right)\ket{1} \\
        R_y(\delta)\ket{1} &= \begin{pmatrix}
            \cos\left(\frac{\delta}{2}\right) & -\sin\left(\frac{\delta}{2}\right) \\ \sin\left(\frac{\delta}{2}\right) & \cos\left(\frac{\delta}{2}\right)
        \end{pmatrix} \begin{pmatrix}
            0 \\ 1
        \end{pmatrix} = \begin{pmatrix}
            -\sin\left(\frac{\delta}{2}\right) \\ \cos\left(\frac{\delta}{2}\right)
        \end{pmatrix} = -\sin\left(\frac{\delta}{2}\right)\ket{0} + \cos\left(\frac{\delta}{2}\right)\ket{1}
    \end{split}
\end{equation*}

And finally

\begin{equation*}
    \begin{split}
        R_x(\delta)\ket{0} &= \begin{pmatrix}
            \cos\left(\frac{\delta}{2}\right) & -i\sin\left(\frac{\delta}{2}\right) \\ -i\sin\left(\frac{\delta}{2}\right) & \cos\left(\frac{\delta}{2}\right)
        \end{pmatrix} \begin{pmatrix}
            1 \\ 0
        \end{pmatrix} = \begin{pmatrix}
            \cos\left(\frac{\delta}{2}\right) \\ -i\sin\left(\frac{\delta}{2}\right)
        \end{pmatrix} = \cos\left(\frac{\delta}{2}\right)\ket{0} - i\sin\left(\frac{\delta}{2}\right)\ket{1} \\
        R_x(\delta)\ket{1} &= \begin{pmatrix}
            \cos\left(\frac{\delta}{2}\right) & -i\sin\left(\frac{\delta}{2}\right) \\ -i\sin\left(\frac{\delta}{2}\right) & \cos\left(\frac{\delta}{2}\right)
        \end{pmatrix} \begin{pmatrix}
            0 \\ 1
        \end{pmatrix} = \begin{pmatrix}
            -i\sin\left(\frac{\delta}{2}\right) \\ \cos\left(\frac{\delta}{2}\right)
        \end{pmatrix} = -i\sin\left(\frac{\delta}{2}\right)\ket{0} + \cos\left(\frac{\delta}{2}\right)\ket{1}
    \end{split}
\end{equation*}

Note that since $R_z$ is a rotation around the $Z$ axis and since $\ket{0}$ and $\ket{1}$ are on this same axis, these states remain invariant under this gate. On the other hand, the gates $R_y$ and $R_x$ leave them in a state of superposition.

### Rotation Gates with respect to an arbitrary axis

Once we have seen how rotations are with respect to the main axes $X$, $Y$ and $Z$, let's see what the rotation of a qubit in the Bloch sphere looks like with respect to any arbitrary axis and by any angle.

### <a name="remark_5_7">Remark 5.7</a>

Let $n$ be a vector in $\mathbb{R}^3$ with $n = (n_x,n_y,n_z)$, that is, what this vector gives, is a direction in $\mathbb{R}^3$, then the rotation around that axis $n$ by an angle $\delta$ is given by

\begin{equation*}
    R_n(\delta) = e^{-i\frac{\delta}{2} (n_x\hat{\sigma_x} + n_y\hat{\sigma_y} + n_z\hat{\sigma_z})}
\end{equation*}

Since $\hat{\sigma_x}^2 = \hat{\sigma_y}^2 = \hat{\sigma_z}^2 = I$, any linear combination of these operators has this same property and therefore we can use **Theorem 5.5**, like this

\begin{equation*}
    R_n(\delta) = cos\left(\frac{\delta}{2}\right)I - i\sin\left(\frac{\delta}{2}\right) (n_x\hat{\sigma_x} + n_y\hat{\sigma_y} + n_z\hat{\sigma_z})
\end{equation*}


### Example

We can see the Hadamard gate in its matrix form and note it can be decomposed as follows

\begin{equation*}
    \begin{split}
        H = \frac{1}{\sqrt{2}}\begin{pmatrix}1 & 1 \\ 1 & -1\end{pmatrix} = \frac{1}{\sqrt{2}} \left[ \begin{pmatrix}0 & 1 \\ 1 & 0\end{pmatrix} + \begin{pmatrix}1 & 0 \\ 0 & -1\end{pmatrix} \right] = \frac{1}{\sqrt{2}}(\hat{\sigma_x} + \hat{\sigma_z})
    \end{split}
\end{equation*}

where $n = \left(\frac{1}{\sqrt{2}}, 0, \frac{1}{\sqrt{2}}\right)$. Therefore

\begin{equation*}
    \begin{split}
        R_n(\pi) &= cos\left(\frac{\pi}{2}\right)I - i\sin\left(\frac{\pi}{2}\right)H \\
        &= 0I - iH = e^{-i\frac{\pi}{2}}H \cong H
    \end{split}
\end{equation*}

since $e^{-i\frac{\pi}{2}}$ is a global phase. The geometric interpretation we can give to this fact, is that the Hadamard gate is essentially a rotation by an angle $\pi$ around the axis $n = \left(\frac{1}{\sqrt{2}}, 0, \frac{1}{\sqrt{2}}\right)$ that lies on the $XZ$ plane. That is, **the Hadamard gate is in fact a very particular rotational gate**.
